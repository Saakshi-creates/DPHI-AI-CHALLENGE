```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Training_set_begs.csv')

data=pd.DataFrame(data)

data.head()
```

```python
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Testing_set_begs.csv')
```

## DATA CLEANING

```python
X=data.drop('Survived_1_year',axis=1)
```

```python
y=data['Survived_1_year']
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=1,stratify=y)
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
X_train.isnull().sum()
```

```python
m=X_train["Treated_with_drugs"].mode()
m
```

```python
#X_train.loc[X_train.Treated_with_drugs == X_train['Treated_with_drugs'].isnull(),'Treated_with_drugs']='DX6'
X_train['Treated_with_drugs']=X_train['Treated_with_drugs'].fillna(X_train["Treated_with_drugs"].mode()[0])
```

```python
X_train['A'].fillna(X_train['A'].mode()[0],inplace=True)
X_train['B'].fillna(X_train['B'].mode()[0],inplace=True)
X_train['C'].fillna(X_train['C'].mode()[0],inplace=True)
X_train['D'].fillna(X_train['D'].mode()[0],inplace=True)
X_train['E'].fillna(X_train['E'].mode()[0],inplace=True)
X_train['F'].fillna(X_train['F'].mode()[0],inplace=True)
X_train['Z'].fillna(X_train['Z'].mode()[0],inplace=True)
```

```python
X_train['Number_of_prev_cond']=X_train['Number_of_prev_cond'].fillna(X_train['Number_of_prev_cond'].mode()[0])
```

```python
X_train.info()
```

```python
X_train.isnull().sum()
```

```python
X_train.info()
```

```python
drugs=X_train['Treated_with_drugs'].str.get_dummies(sep=' ')
drugs.head()
```

```python
data=pd.concat([X_train,drugs],axis=1)
data=data.drop('Treated_with_drugs',axis=1)
data.loc[data.Patient_Smoker == 'Cannot say','Patient_Smoker']='NO'
```

```python
data.head()
```

```python
data.info()
```

```python
data=data.drop('Patient_mental_condition',axis=1)
```

```python
data=pd.get_dummies(data,columns=['Patient_Rural_Urban','Patient_Smoker'])
data.head()
```

```python
data.drop('ID_Patient_Care_Situation',axis=1,inplace=True)
```

```python
data.head()
```
